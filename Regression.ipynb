Loading and Preprocessing

```python
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

```

```python
california_housing = fetch_california_housing()
print(california_housing);
```

```python
df = pd.DataFrame(california_housing.data,columns=california_housing.feature_names)
print(df);
```

```python
print("Dataset shape:", df.shape)
```

```python
df['MedHouseVal'] = california_housing.target
df
```

```python
print(df.isnull().sum())
```

There is no missing values.

```python
df.describe()
```

```python
numberscol=df.select_dtypes(include=['int64', 'float64']).columns
for col in numberscol:
   q1 = df[col].quantile(0.25)
   q3 = df[col].quantile(0.75)
   iqr = q3 - q1
   lower_bound = q1 - 1.5 * iqr
   upper_bound = q3 + 1.5 * iqr
   outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
   print(f"Outliers: {len(outliers)}")
   df[col] = df[col].clip(lower_bound, upper_bound)
   df[col]
```

```python
from sklearn.preprocessing import StandardScaler
X = df.drop('MedHouseVal', axis=1)
y = df['MedHouseVal']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
print(X_scaled)
```

```python
df = pd.concat([X_scaled, y], axis=1)
df
```

first load the California Housing data set from sklearn.datasets and convert to dataframe.  then add the target colum , Then check the null values using df.isnull.sum() there is no null values because missing values negatively effect the model traing . Identify the numerical coloum and  calculate the IQR and capped the outlier values  , Outliers causes to make the errors especially in the case of linear regression.then apply StandardScaler()  which convert mean=0 and standard deviation to 1 through this treat all the features equally  .

Regression Algorithm Implementation
(1)Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(random_state=42),
    'Random Forest': RandomForestRegressor(random_state=42, n_estimators=100)
}
```

```python
lr_model = models['Linear Regression']
lr_model.fit(X_train, y_train)
lr_pred = lr_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse =mean_squared_error(y_test, lr_pred)
print("Mean Square error:", mse)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mas = mean_absolute_error(y_test, lr_pred)
print("Mean Absolute error:", mas)
```

```python
r2_score = lr_model.score(X_test, y_test)
print("R2 Score:", r2_score)
```

Decision Tree

```python
dt_model = models['Decision Tree']
dt_model.fit(X_train, y_train)
dt_pred = dt_model.predict(X_test)
```

```python
mse =mean_squared_error(y_test, dt_pred)
print("Mean Square error:", mse)
```

```python
mae = mean_absolute_error(y_test, dt_pred)
print("Mean Absolute error:", mae)
```

```python
r2_score = dt_model.score(X_test, y_test)
print("R2 Score:", r2_score)
```

Random Forest Tree

```python
rf_model = models['Random Forest']
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
```

```python
mse=mean_squared_error(y_test, rf_pred)
print("Mean Square error:", mse)
```

```python
mae = mean_absolute_error(y_test, rf_pred)
print("Mean Absolute error:", mae)
```

```python
r2_score = rf_model.score(X_test, y_test)
print("R2 Score:", r2_score)
```

Linear Regression:
The Linear Regression model has an MSE of 0.528, MAE of 0.533, and R2 of 0.601. This means it explains about 60% of the variation in house prices. The model provides a reasonable baseline but is limited because it cannot capture the nonlinear relationships within the housing data.

Decision Tree Regressor:
The Decision Tree Regressor shows an MSE of 0.468, MAE of 0.441, and R2 of 0.646. Its R2 is higher than the Linear Regression model, indicating a better fit. While it captures nonlinear patterns effectively, its moderate MSE suggests it may be overfitting the training data, which limits its generalization capability.

Random Forest Regressor:
The Random Forest Regressor has an MSE of 0.321, MAE of 0.361, and R2 of 0.757, explaining over 75% of the variance in house prices. Both the MSE and MAE are significantly lower than the other models, meaning its predictions are the most accurate. It successfully captures complex patterns while avoiding overfitting through its ensemble method.

Conclusion: The Random Forest Regressor is the best-performing model.
